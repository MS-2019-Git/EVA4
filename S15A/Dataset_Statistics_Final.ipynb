```
#calculate STd and Mean
from tqdm import notebook
from PIL import Image
import numpy as np
def mean_std(path):

    n = 0
    s = np.zeros(3)
    sq = np.zeros(3)
    start = 0
    stop = 80000
    for part in range(5):
      print(f"from {start} to {stop} of set {part+1}")
      for i in notebook.tqdm(range(start,stop)):
          
            img = Image.open(f'/content/dataset{path}{str(i+1)}.jpg')
            x = np.array(img)/255
            s += x.sum(axis=(0,1))
            sq += np.sum(np.square(x), axis=(0,1))
            n += x.shape[0]*x.shape[1]
      start += 80000
      stop+=80000


    mu = s/n
    std = np.sqrt((sq/n - np.square(mu)))
    print(mu, sq/n, std, n)
```
